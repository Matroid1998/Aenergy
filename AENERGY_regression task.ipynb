# Linear Regression

```python
##train linear reg on 2 datasets and its poly features
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import RepeatedKFold
cv = RepeatedKFold(n_splits=4, n_repeats=5, random_state=42)
```

```python
from sklearn.model_selection import cross_validate
```

```python
Xs = [X_train_eo_std,X_train_eo_std2,X_train_eo_std3,X_train_eo_std4]
Xs2 = [X_train_eos,X_train_eos2,X_train_eos3,X_train_eos4]
```

```python
lin_val = []
lin_train = []
i=1
for X in Xs:
  lin_reg = LinearRegression()
  score = cross_validate(lin_reg,X,y_train,cv = cv , scoring ='neg_root_mean_squared_error',return_train_score=True)
  print("degree = {}".format(i))
  print('validation score : ')
  print(np.mean(score['test_score']))
  print('train score')
  print(np.mean(score['train_score']))
  i+=1
  lin_val.append(np.mean(score['test_score']))
  lin_train.append(np.mean(score['train_score']))
```

```python
lin_val2=[]
lin_train2=[]
i=1
for X in Xs2:
  lin_reg = LinearRegression()
  score = cross_validate(lin_reg,X,y_train,cv = cv , scoring ='neg_root_mean_squared_error',return_train_score=True)
  print("degree = {}".format(i))
  print('validation score : ')
  print(np.mean(score['test_score']))
  print('train score')
  print(np.mean(score['train_score']))
  i+=1
  lin_val2.append(np.mean(score['test_score']))
  lin_train2.append(np.mean(score['train_score']))
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[lin_val,lin_val2],labels=['validation on 1st dataset ','validation on 2nd dataset'])
plt.show()
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[lin_train,lin_train2],labels=['train score on 1st dataset ','train score on 2nd dataset'])
plt.show()
```

# Ridge Regression


```python
##train ridge regression on both datasets and its poly features
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
params_grid = [{
    'alpha' : [0.1,1,10,50,100,150,200,1000]
}]
i=1
ridge_val = []
ridge_train = [] 
for X in Xs:
  ridge = Ridge()
  ridge_GS = GridSearchCV(ridge,params_grid,cv = cv , scoring = 'neg_root_mean_squared_error', return_train_score=True)
  ridge_GS.fit(X,y_train)
  print('degree : {}'.format(i))
  print(ridge_GS.best_score_)  
  print(ridge_GS.best_params_)
  ridge_result = ridge_GS.cv_results_
  print("train results")
  for mean_score, params in zip(ridge_result["mean_train_score"],ridge_result["params"]):
      print(-mean_score, params)
  print("test results")    
  for mean_score, params in zip(ridge_result["mean_test_score"],ridge_result["params"]):
    print(-mean_score, params)
  i+=1      
  print('----------------------------------------------------------')  
  ridge_val.append(ridge_result['mean_test_score'])
  ridge_train.append(ridge_result['mean_train_score'])
```

```python
ridge_val2=[]
ridge_train2=[]
i=1
for X in Xs2:
  ridge = Ridge()
  ridge_GS = GridSearchCV(ridge,params_grid,cv = cv , scoring = 'neg_root_mean_squared_error', return_train_score=True)
  ridge_GS.fit(X,y_train)
  print('degree : {}'.format(i))
  print(ridge_GS.best_score_)  
  print(ridge_GS.best_params_)
  ridge_result = ridge_GS.cv_results_
  print("train results")
  for mean_score, params in zip(ridge_result["mean_train_score"],ridge_result["params"]):
      print(-mean_score, params)
  print("test results")    
  for mean_score, params in zip(ridge_result["mean_test_score"],ridge_result["params"]):
    print(-mean_score, params)
  i+=1      
  print('----------------------------------------------------------')  
  ridge_val2.append(ridge_result['mean_test_score'])
  ridge_train2.append(ridge_result['mean_train_score'])  
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[ridge_val[0],ridge_val2[0]],labels=['validation on 1st dataset ','validation on 2nd dataset'])
plt.show()
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[ridge_train[0],ridge_train2[0]],labels=['train on 1st dataset ','train on 2nd dataset'])
plt.show()
```
