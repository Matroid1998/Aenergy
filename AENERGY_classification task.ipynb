```python
###fill f21 with different methods and plot them
from matplotlib.pyplot import figure

figure(figsize=(20, 15), dpi=80)
plt.subplot(2, 4, 1)
plt.plot(f21_train)
plt.title('f21')
mean_f21 = f21_train.fillna(f21_train.mean())
plt.subplot(2, 4, 2)
plt.plot(mean_f21)
plt.title('mean')
linear_f21 = f21_train.interpolate(method='linear')
linear_f21 = linear_f21.fillna(linear_f21.mean())
plt.subplot(2, 4, 3)
plt.plot(linear_f21)
plt.title('linear')
nearest_f21 = f21_train.interpolate(method='nearest')
plt.subplot(2, 4, 4)
plt.plot(nearest_f21)
plt.title('nearest')
poly2_f21 = f21_train.interpolate(method='polynomial', order=2)
plt.subplot(2, 4, 5)
plt.plot(poly2_f21)
plt.title('poly2')
poly3_f21 = f21_train.interpolate(method='polynomial', order=2)
plt.subplot(2, 4, 6)
plt.plot(poly3_f21)
plt.title('poly3')
derivative_f21 = f21_train.interpolate(method='from_derivatives')
plt.subplot(2, 4,7)
plt.plot(derivative_f21)
plt.title('derivatives')
spline_f21 = f21_train.interpolate(method= 'spline',order = 2)
plt.subplot(2, 4, 8)
plt.plot(spline_f21)
plt.title('spline')
plt.show()
```

# Decision Tree

```python
##train DT on various hyperparameters and 2 dataset that obtained after preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
```

```python
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
DF = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
params_grid=[{'min_samples_leaf':[1,5,10,15,20,25,50,75,100],
              'max_depth':[None,3,4,5,6,7,10,15]
              ,'min_samples_split':[2,10,20,30,50,100]}]
```

```python
DF_GS =GridSearchCV (DF,params_grid,cv = 4,scoring =['accuracy','precision','recall','f1'],return_train_score = True,refit='accuracy')
```

```python
DF_GS.fit(X_train_,y_train)
```

```python
DF_GS.best_params_
```

```python
DF_GS.best_score_
```

```python
DF_GS.cv_results_.keys()
```

```python
DF_result = DF_GS.cv_results_
```

```python
for mean_score,recall_score,precision_score,f1_score,params in zip(DF_result["mean_test_accuracy"],
                                                                   DF_result['mean_test_recall'],DF_result['mean_test_precision']
                                                                   ,DF_result["mean_test_f1"],DF_result["params"]):
  print('mean_accuracy',"%.4f"%mean_score,'mean recall',"%.4f"%recall_score,'mean precision',"%.4f"%precision_score,
        'mean_f1',"%.4f"%f1_score, params)
```

```python
for mean_score,recall_score,precision_score,f1_score,params in zip(DF_result["mean_train_accuracy"],
                                                                   DF_result['mean_train_recall'],DF_result['mean_train_precision']
                                                                   ,DF_result["mean_train_f1"],DF_result["params"]):
  print('mean_accuracy',"%.4f"%mean_score,'mean recall',"%.4f"%recall_score,'mean precision',"%.4f"%precision_score,
        'mean_f1',"%.4f"%f1_score, params)
```

# Best estimators

```python
##here we want to choose between DT and RF
from sklearn.model_selection import cross_validate
DF_best = DecisionTreeClassifier(max_depth=5,min_samples_leaf = 5,min_samples_split = 20)
```

```python
cv = RepeatedKFold(n_splits=5, n_repeats=10, random_state=42)
```

```python
score_DF = cross_validate(DF_best,X_train_std,y_train,cv = cv , return_train_score = True , return_estimator= True,scoring=['accuracy','recall'])
```

```python
score_DF
```

```python
RF_best = RandomForestClassifier(bootstrap = False, max_depth = None, max_features=13,min_samples_leaf = 5)
```

```python
score_RF = cross_validate(RF_best,X_train_std,y_train,cv = cv , return_train_score = True , return_estimator= True,scoring=['accuracy','recall'])
```

```python
score_RF
```

```python
##plot train and validation results for DT
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_DF['train_accuracy'],score_DF['test_accuracy']],labels=['train accuracy DT','test accuracy DT'])
plt.show()
```

```python
##plot train and validation accuracy for RF
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_RF['train_accuracy'],score_RF['test_accuracy']],labels=['train accuracy RF','test accuracy RF'])
plt.show()
```

```python
##plot train accuracy for RF and DT together

fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_DF['train_accuracy'],score_RF['train_accuracy']],labels=['train accuracy DT','train accuracy RF'])
plt.show()
```

```python
##plot test accuracy for RF and DT together
fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_DF['test_accuracy'],score_RF['test_accuracy']],labels=['test accuracy DT','test accuracy RF'])
plt.show()
```

```python
##plot train recall for RF and DT together

fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_DF['train_recall'],score_RF['train_recall']],labels=['train Recall DT','train recall RF'])
plt.show()
```

```python
##plot test recall for RF and DT together

fig = plt.figure(figsize =(10, 7))
plt.boxplot(x=[score_DF['test_recall'],score_RF['test_recall']],labels=['test racall DT','test racall RF'])
plt.show()
```

```python
##obtain best estimator on RF based on validation recall
iopt = np.argmax(score_RF['test_recall'])
```

```python
iopt
```

```python
opt_RF = score_RF["estimator"][iopt]
```
